```python
from diagrams import Diagram, Cluster, Edge
from diagrams.onprem.inmemory import Redis
```

```python
with Diagram('Redis Replication', filename='replication/diagram') as diag:
    with Cluster('midgard'):
        master = Redis('midgard')
        r1 = Redis('midgard-r1')
        r2 = Redis('midgard-r2')
        replica_of = Edge(label='replica of')
        midgard = [r1 - replica_of - master - replica_of - r2]
diag
```

```python
with Diagram('Redis Sentinel', filename='sentinel/diagram') as diag:
    with Cluster('midgard'):
        master = Redis('midgard')
        r1 = Redis('midgard-r1')
        r2 = Redis('midgard-r2')
        replica_of = Edge(label='replica of')
        midgard = [r1 - replica_of - master - replica_of - r2]
    midgard << Edge(label='monitor') << Redis('heimdall')
diag
```
